```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_csv('your_file.csv')


continent_color_map = {
    'Asia': 'red',
    'Europe': 'blue',
    'Africa': 'green',
    'N. America': 'orange',
    'S. America': 'purple',
    'Oceania': 'yellow'
}

data['Color'] = data['Continent'].map(continent_color_map)

data = data.dropna()

lower_threshold = 10
upper_threshold = 100

data_within_threshold = data[(data['all electricity generation'] > lower_threshold)& (data['all electricity generation'] < upper_threshold)]
label_distance_threshold = 6

fig_scatter = go.Figure()

constant_marker_size = 20
fig_scatter.add_trace(go.Scatter(
    x=data_within_threshold['all renewables'],
    y=data_within_threshold['all renewables 2015'],
    mode='markers',
    marker=dict(
        size=constant_marker_size,
        color=data_within_threshold['Color'],
        opacity=0.7,
        line=dict(width=0.5, color='black'),
    ),
    text=data_within_threshold['Country'],
))

annotations = []
for i in range(len(data_within_threshold)):
    x_i = data_within_threshold.iloc[i]['all renewables']
    y_i = data_within_threshold.iloc[i]['all renewables 2015']
    text_i = data_within_threshold.iloc[i]['Country']

    skip_label = False
    for ann in annotations:
        x_dist = abs(x_i - ann['x'])
        y_dist = abs(y_i - ann['y'])
        if x_dist < label_distance_threshold and y_dist < label_distance_threshold:
            skip_label = True
            ann['text'] += f"<br>{text_i}"
            break

    if not skip_label:
        annotations.append(dict(x=x_i, y=y_i, text=text_i,
                                showarrow=False, font=dict(color='black', size=12),
                                xshift=30, yshift=-20))

def adjust_label_positions(annotations):
    for i in range(len(annotations)):
        for j in range(i + 1, len(annotations)):
            label_i = annotations[i]
            label_j = annotations[j]

            if abs(label_i['x'] - label_j['x']) < label_distance_threshold and abs(label_i['y'] - label_j['y']) < label_distance_threshold:
                annotations[j]['y'] -= 15
                annotations[j]['x'] -= 10

adjust_label_positions(annotations)

fig_scatter.update_layout(annotations=annotations)

fig_scatter.update_layout(
    xaxis=dict(title='Difference (Renewables in 2020 - Renewables in 2015)', showgrid=False,),
    yaxis=dict(title='All Renewables', showgrid=False, showline=False),
    plot_bgcolor='rgba(217, 226, 413, 0.8)'
)

fig_scatter.update_layout(width=1000, height=1000)

fig_scatter.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_csv('your_file.csv')

continent_color_map = {
    'Asia': 'red',
    'Europe': 'blue',
    'Africa': 'green',
    'N. America': 'orange',
    'S. America': 'purple',
    'Oceania': 'yellow'
}

continent_avg_renewables = data.groupby('Continent')['all renewables'].mean().reset_index()

fig = go.Figure()

fig.add_trace(go.Bar(
    x=continent_avg_renewables['Continent'],
    y=continent_avg_renewables['all renewables'],
    marker=dict(color=[continent_color_map[cont] for cont in continent_avg_renewables['Continent']]),
    text=continent_avg_renewables['Continent'],
    textposition='inside'
))

fig.update_layout(
    title='Average Renewable Percentage for Each Continent',
    xaxis=dict(title='Continent'),
    yaxis=dict(title='Average Renewable Percentage'),
    bargap=0.1,
    height=500,
    )

fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_csv('your_file.csv')

continent_color_map = {
    'Asia': 'red',
    'Europe': 'blue',
    'Africa': 'green',
    'N. America': 'orange',
    'S. America': 'purple',
    'Oceania': 'yellow'
}

data['Hydropower'] = ((data['all renewables'] - data['excluding hydropower']) / data['all renewables']) * 100

continent_avg_hydropower = data.groupby('Continent')['Hydropower'].mean().reset_index()

fig = go.Figure()

fig.add_trace(go.Bar(
    x=continent_avg_hydropower['Continent'],
    y=continent_avg_hydropower['Hydropower'],
    marker=dict(color=[continent_color_map[cont] for cont in continent_avg_hydropower['Continent']]),
    text=continent_avg_hydropower['Continent'],
    textposition='inside'
))

fig.update_layout(
    title='Average Hydropower Percentage for Each Continent',
    xaxis=dict(title='Continent'),
    yaxis=dict(title='Average Hydropower Percentage'),
    bargap=0.1,
    height=500,
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_csv('your_file.csv')

negative_differences = data[data[' difference'] < 0]

top_10_negative_diff = negative_differences.nsmallest(10, ' difference')

top_10_negative_diff = top_10_negative_diff.sort_values(' difference', ascending=True)

fig = go.Figure()

fig.add_trace(go.Bar(
    y=top_10_negative_diff['Country'],
    x=top_10_negative_diff[' difference'],
    orientation='h',
    marker=dict(color='red'),
    text=top_10_negative_diff['Country'],
    textposition='inside',
))

fig.update_layout(
    title='Top 10 Countries with Largest Absolute Negative Differences',
    xaxis=dict(title='Difference'),
    yaxis=dict(title='Country'),
    bargap=0.1,
    margin=dict(l=150),
    height=500,
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_csv('your_file.csv')

positive_differences = data[data[' difference'] > 0]

top_10_positive_diff = positive_differences.nlargest(10, ' difference')

top_10_positive_diff = top_10_positive_diff.sort_values(' difference', ascending=False)

fig = go.Figure()

fig.add_trace(go.Bar(
    y=top_10_positive_diff['Country'],
    x=top_10_positive_diff[' difference'],
    orientation='h',
    marker=dict(color='blue'),
    text=top_10_positive_diff['Country'],
    textposition='inside',
))

fig.update_layout(
    title='Top 10 Countries with Largest Absolute Positive Differences',
    xaxis=dict(title='Difference'),
    yaxis=dict(title='Country'),
    bargap=0.1,
    margin=dict(l=150),
    height=500,
)

fig.show()
```
